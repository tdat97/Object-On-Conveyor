```python
import numpy as np
import cv2
from utils import tool
import time

def auto_light_cam(cam, offset=0.02):
    before_light_on = False
    light_on = False
    before_ratio = 0
    before_img = np.zeros(cam.img_shape, dtype=np.uint8)
    test_marker = ""
    test_marker2 = ""
    
    for i in range(1000):
        time.sleep(0.01)
        
        # 촬영
        img = cam.get_image()
        if img is None: continue
        
        
        # 동작 감지에 따른 조명 동작 # 2연속 조건 만족해야 동작
        diff_img = tool.get_diff_img(before_img, img)
        ratio = tool.diff2ratio(diff_img)
        
        
        # 조명 켜져있을때만 보내기
        if light_on and before_light_on:
            test_marker = "S"
            print('S', end='')
        else:
            test_marker = ""

        if not light_on and ratio > offset and before_ratio > offset:
            light_on = True
            test_marker2 = "O"
            print('O', end='')
            time.sleep(0.15)
        if light_on and ratio <= offset and before_ratio <= offset:
            light_on = False
            test_marker2 = "X"
            print('X', end='')
            time.sleep(0.15)
            
            
        # img_copy = cv2.resize(img, dsize=(0,0), fx=0.4, fy=0.4)
        img_copy = cv2.resize(img, dsize=(0,0), fx=1, fy=1)
        cv2.putText(img_copy, f"Ratio : {ratio:.2f}", (0, 100), cv2.FONT_HERSHEY_SIMPLEX, 
                    fontScale=1, color=(255,0,0), 
                    thickness=2, lineType=cv2.LINE_AA)
        cv2.putText(img_copy, f"send : {test_marker}", (0, 200), cv2.FONT_HERSHEY_SIMPLEX, 
                    fontScale=1, color=(255,0,0), 
                    thickness=2, lineType=cv2.LINE_AA)
        cv2.putText(img_copy, f"light : {test_marker2}", (0, 300), cv2.FONT_HERSHEY_SIMPLEX, 
                    fontScale=1, color=(255,0,0), 
                    thickness=2, lineType=cv2.LINE_AA)
        
        diff_img = cv2.merge([diff_img, diff_img, diff_img])
        new_img = np.hstack([img_copy, diff_img])
        
        cv2.imshow('test', new_img)
        if cv2.waitKey(1) & 0xff == ord('q'): return
    
    
        print(" ", end='')

        before_light_on = light_on
        before_img = img
        before_ratio = ratio


        
        

        
import serial
import time

myserial = serial.Serial("COM5", 9600, timeout=0, bytesize=serial.EIGHTBITS, 
                                  stopbits=serial.STOPBITS_ONE, parity=serial.PARITY_ODD)


myserial.write(b'\xA0\x01\x01\xA2')

auto_light_cam(cam)
cv2.destroyAllWindows()

myserial.write(b'\xA0\x01\x00\xA1')
myserial.close()
```

```python
def get_diff_img(img1, img2):
    assert img1.shape == img2.shape
    
    # To Grayscale
    if len(img1.shape) == 3:
        img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

    # subtract
    sub = img1.astype(np.int32) - img2.astype(np.int32)
    result = np.abs(sub).astype(np.uint8)
    _, result = cv2.threshold(result, 5,255, cv2.THRESH_BINARY)
    
    # dilate
    # kernel = np.ones((3,3))
    kernel = np.ones((5,5))
    result = cv2.erode(result, kernel, iterations=1)
    result = cv2.dilate(result, kernel, iterations=2)
    return result

aaa = get_diff_img(img1, img2)
plt.imshow(aaa)
```
